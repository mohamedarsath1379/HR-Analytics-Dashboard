###  Business Problem
Employee attrition costs companies **33% of an employee's annual salary** in replacement costs.  
HR teams need data-driven insights to identify **why employees leave** and **who is at risk** of leaving.

1. What is the overall attrition rate?
2. Which department has the highest attrition?
3. Does overtime cause attrition?
4. Which salary band has highest exit rate?
5. Who are the high-risk employees?

##  Step 5: Feature Engineering — Add New Columns

```python
%%sql
-- Add Age Group column
ALTER TABLE hr_data ADD COLUMN AgeGroup VARCHAR(20);
```

```python
%%sql
-- Update Age Group values
UPDATE hr_data SET AgeGroup =
    CASE
        WHEN Age BETWEEN 18 AND 25 THEN '18-25'
        WHEN Age BETWEEN 26 AND 35 THEN '26-35'
        WHEN Age BETWEEN 36 AND 45 THEN '36-45'
        WHEN Age BETWEEN 46 AND 55 THEN '46-55'
        ELSE '55+'
    END;
```

```python
%%sql
-- Add Salary Band column
ALTER TABLE hr_data ADD COLUMN SalaryBand VARCHAR(20);
```

```python
%%sql
-- Update Salary Band values
UPDATE hr_data SET SalaryBand =
    CASE
        WHEN MonthlyIncome < 3000                  THEN 'Low (<3K)'
        WHEN MonthlyIncome BETWEEN 3000 AND 6000   THEN 'Medium (3K-6K)'
        WHEN MonthlyIncome BETWEEN 6001 AND 10000  THEN 'High (6K-10K)'
        ELSE 'Very High (>10K)'
    END;
```

```python
%%sql
-- Add Experience Band column
ALTER TABLE hr_data ADD COLUMN ExperienceBand VARCHAR(20);
```

```python
%%sql
-- Update Experience Band values
UPDATE hr_data SET ExperienceBand =
    CASE
        WHEN TotalWorkingYears BETWEEN 0  AND 5  THEN '0-5 Years'
        WHEN TotalWorkingYears BETWEEN 6  AND 10 THEN '6-10 Years'
        WHEN TotalWorkingYears BETWEEN 11 AND 20 THEN '11-20 Years'
        ELSE '20+ Years'
    END;
```

```python
%%sql
-- Add Education Label column
ALTER TABLE hr_data ADD COLUMN EducationLabel VARCHAR(20);
```

```python
%%sql
-- Update Education Label values
UPDATE hr_data SET EducationLabel =
    CASE Education
        WHEN 1 THEN 'Below College'
        WHEN 2 THEN 'College'
        WHEN 3 THEN 'Bachelor'
        WHEN 4 THEN 'Master'
        WHEN 5 THEN 'Doctor'
    END;
```

```python
%%sql
-- Add Attrition Risk column
ALTER TABLE hr_data ADD COLUMN AttritionRisk VARCHAR(15);
```

```python
%%sql
-- Update Attrition Risk values
UPDATE hr_data SET AttritionRisk =
    CASE
        WHEN JobSatisfaction <= 2
         AND WorkLifeBalance <= 2
         AND OverTime = 'Yes'  THEN 'HIGH RISK'
        WHEN JobSatisfaction = 3
         AND OverTime = 'Yes'  THEN 'MEDIUM RISK'
        ELSE 'LOW RISK'
    END;
```

```python
%%sql
-- Verify all new columns added
SELECT
    EmployeeNumber,
    Age, AgeGroup,
    MonthlyIncome, SalaryBand,
    TotalWorkingYears, ExperienceBand,
    EducationLabel,
    AttritionRisk
FROM hr_data
LIMIT 10;
```

##  Step 6: KPI Analysis

```python
%%sql
-- Overall KPI Summary
SELECT
    COUNT(*)                                                                      AS Total_Employees,
    SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)                            AS Total_Attrition,
    SUM(CASE WHEN Attrition='No'  THEN 1 ELSE 0 END)                            AS Active_Employees,
    ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2)    AS Attrition_Rate_Pct,
    ROUND(AVG(Age),1)                                                             AS Avg_Age,
    ROUND(AVG(MonthlyIncome),0)                                                  AS Avg_Monthly_Income,
    ROUND(AVG(JobSatisfaction),2)                                                AS Avg_Job_Satisfaction,
    ROUND(AVG(YearsAtCompany),1)                                                 AS Avg_Tenure_Years,
    SUM(CASE WHEN OverTime='Yes' THEN 1 ELSE 0 END)                             AS Total_OT_Employees
FROM hr_data;
```

##  Step 7: Attrition Analysis

```python
%%sql
-- Attrition by Department
SELECT
    Department,
    COUNT(*)                                                                      AS Total,
    SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)                            AS Attrition_Count,
    SUM(CASE WHEN Attrition='No'  THEN 1 ELSE 0 END)                            AS Active_Count,
    ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2)    AS Attrition_Rate
FROM hr_data
GROUP BY Department
ORDER BY Attrition_Rate DESC;
```

```python
%%sql
-- Attrition by Job Role
SELECT
    JobRole,
    COUNT(*)                                                                      AS Total,
    SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)                            AS Attrition_Count,
    ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2)    AS Attrition_Rate,
    ROUND(AVG(MonthlyIncome),0)                                                  AS Avg_Salary
FROM hr_data
GROUP BY JobRole
ORDER BY Attrition_Rate DESC;
```

```python
%%sql
-- Attrition by Age Group
SELECT
    AgeGroup,
    COUNT(*)                                                                      AS Total,
    SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)                            AS Attrition_Count,
    ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2)    AS Attrition_Rate
FROM hr_data
GROUP BY AgeGroup
ORDER BY AgeGroup;
```

```python
%%sql
-- Attrition by Gender
SELECT
    Gender,
    COUNT(*)                                                                      AS Total,
    SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)                            AS Attrition_Count,
    ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2)    AS Attrition_Rate
FROM hr_data
GROUP BY Gender;
```

```python
%%sql
-- Attrition by Salary Band
SELECT
    SalaryBand,
    COUNT(*)                                                                      AS Total,
    SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)                            AS Attrition_Count,
    ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2)    AS Attrition_Rate
FROM hr_data
GROUP BY SalaryBand
ORDER BY Attrition_Rate DESC;
```

```python
%%sql
-- Attrition by OverTime (KEY BUSINESS INSIGHT)
SELECT
    OverTime,
    COUNT(*)                                                                      AS Total,
    SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)                            AS Attrition_Count,
    ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2)    AS Attrition_Rate
FROM hr_data
GROUP BY OverTime;
```

```python
%%sql
-- Attrition by Marital Status
SELECT
    MaritalStatus,
    COUNT(*)                                                                      AS Total,
    SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)                            AS Attrition_Count,
    ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2)    AS Attrition_Rate
FROM hr_data
GROUP BY MaritalStatus
ORDER BY Attrition_Rate DESC;
```

```python
%%sql
-- Attrition by Business Travel
SELECT
    BusinessTravel,
    COUNT(*)                                                                      AS Total,
    SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)                            AS Attrition_Count,
    ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2)    AS Attrition_Rate
FROM hr_data
GROUP BY BusinessTravel
ORDER BY Attrition_Rate DESC;
```

##  Step 8: Satisfaction Analysis

```python
%%sql
-- Satisfaction Scores by Department
SELECT
    Department,
    ROUND(AVG(JobSatisfaction),2)           AS Avg_Job_Satisfaction,
    ROUND(AVG(EnvironmentSatisfaction),2)   AS Avg_Env_Satisfaction,
    ROUND(AVG(RelationshipSatisfaction),2)  AS Avg_Relationship_Satisfaction,
    ROUND(AVG(WorkLifeBalance),2)           AS Avg_WorkLife_Balance
FROM hr_data
GROUP BY Department;
```

```python
%%sql
-- Job Satisfaction vs Attrition
SELECT
    JobSatisfaction,
    COUNT(*)                                                                      AS Total,
    SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)                            AS Attrition_Count,
    ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2)    AS Attrition_Rate
FROM hr_data
GROUP BY JobSatisfaction
ORDER BY JobSatisfaction;
```

```python
%%sql
-- Work Life Balance vs Attrition
SELECT
    WorkLifeBalance,
    COUNT(*)                                                                      AS Total,
    SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)                            AS Attrition_Count,
    ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2)    AS Attrition_Rate
FROM hr_data
GROUP BY WorkLifeBalance
ORDER BY WorkLifeBalance;
```

##  Step 9: Salary Analysis

```python
%%sql
-- Salary by Department and Job Role
SELECT
    Department,
    JobRole,
    COUNT(*)                         AS Headcount,
    ROUND(AVG(MonthlyIncome),0)     AS Avg_Salary,
    MIN(MonthlyIncome)              AS Min_Salary,
    MAX(MonthlyIncome)              AS Max_Salary,
    ROUND(AVG(PercentSalaryHike),1) AS Avg_Hike_Pct
FROM hr_data
GROUP BY Department, JobRole
ORDER BY Department, Avg_Salary DESC;
```

```python
%%sql
-- Performance Rating vs Salary Hike
SELECT
    PerformanceRating,
    COUNT(*)                         AS Headcount,
    ROUND(AVG(PercentSalaryHike),1) AS Avg_Hike_Pct,
    ROUND(AVG(MonthlyIncome),0)     AS Avg_Salary
FROM hr_data
GROUP BY PerformanceRating
ORDER BY PerformanceRating;
```

##  Step 10: Advanced Analysis

```python
%%sql
-- High Risk Employees Table
SELECT
    EmployeeNumber,
    Age,
    Department,
    JobRole,
    MonthlyIncome,
    JobSatisfaction,
    WorkLifeBalance,
    OverTime,
    YearsAtCompany,
    YearsSinceLastPromotion,
    Attrition,
    AttritionRisk
FROM hr_data
WHERE AttritionRisk IN ('HIGH RISK','MEDIUM RISK')
ORDER BY AttritionRisk, JobSatisfaction limit 10;
```

```python
%%sql
-- Departments Above Company Average Attrition (Using CTE)
WITH company_avg AS (
    SELECT ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2) AS avg_rate
    FROM hr_data
)
SELECT
    h.Department,
    COUNT(*)                                                                      AS Total,
    ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2)    AS Dept_Rate,
    c.avg_rate                                                                    AS Company_Avg,
    CASE
        WHEN ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2) > c.avg_rate
        THEN '⚠ ACTION NEEDED'
        ELSE '✅ HEALTHY'
    END AS Status
FROM hr_data h, company_avg c
GROUP BY h.Department, c.avg_rate
ORDER BY Dept_Rate DESC;
```

```python
%%sql
-- Promotion Gap Analysis
SELECT
    Department,
    JobRole,
    ROUND(AVG(YearsSinceLastPromotion),1)            AS Avg_Years_No_Promotion,
    COUNT(*)                                          AS Headcount,
    SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END) AS Attrition_Count
FROM hr_data
GROUP BY Department, JobRole
HAVING Avg_Years_No_Promotion > 3
ORDER BY Avg_Years_No_Promotion DESC;
```

```python
%%sql
-- Training Impact on Attrition
SELECT
    TrainingTimesLastYear,
    COUNT(*)                                                                      AS Total,
    SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)                            AS Attrition_Count,
    ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2)    AS Attrition_Rate
FROM hr_data
GROUP BY TrainingTimesLastYear
ORDER BY TrainingTimesLastYear;
```

##  Step 11: Create SQL Views for Power BI

```python
%%sql
-- View 1: Main Dashboard View (Connect this to Power BI)
CREATE OR REPLACE VIEW vw_hr_main AS
SELECT
    EmployeeNumber, Age, AgeGroup, Gender, MaritalStatus,
    Department, JobRole, JobLevel,
    MonthlyIncome, SalaryBand, PercentSalaryHike,
    JobSatisfaction, EnvironmentSatisfaction,
    RelationshipSatisfaction, WorkLifeBalance,
    OverTime, BusinessTravel, DistanceFromHome,
    TotalWorkingYears, ExperienceBand,
    YearsAtCompany, YearsInCurrentRole,
    YearsSinceLastPromotion, YearsWithCurrManager,
    TrainingTimesLastYear, PerformanceRating,
    EducationLabel, EducationField,
    Attrition, AttritionRisk
FROM hr_data;
```

```python
%%sql
-- View 2: KPI Summary View
CREATE OR REPLACE VIEW vw_kpi_summary AS
SELECT
    COUNT(*)                                                                      AS Total_Employees,
    SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)                            AS Total_Attrition,
    SUM(CASE WHEN Attrition='No'  THEN 1 ELSE 0 END)                            AS Active_Employees,
    ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2)    AS Attrition_Rate,
    ROUND(AVG(Age),1)                                                             AS Avg_Age,
    ROUND(AVG(MonthlyIncome),0)                                                  AS Avg_Income,
    ROUND(AVG(JobSatisfaction),2)                                                AS Avg_Satisfaction,
    ROUND(AVG(YearsAtCompany),1)                                                 AS Avg_Tenure,
    SUM(CASE WHEN OverTime='Yes' THEN 1 ELSE 0 END)                             AS OT_Employees
FROM hr_data;
```

```python
%%sql
-- View 3: Department Summary View
CREATE OR REPLACE VIEW vw_dept_summary AS
SELECT
    Department,
    COUNT(*)                                                                      AS Total_Employees,
    SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)                            AS Attrition_Count,
    ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2)    AS Attrition_Rate,
    ROUND(AVG(MonthlyIncome),0)                                                  AS Avg_Salary,
    ROUND(AVG(JobSatisfaction),2)                                                AS Avg_Satisfaction,
    ROUND(AVG(WorkLifeBalance),2)                                                AS Avg_WLB,
    SUM(CASE WHEN OverTime='Yes' THEN 1 ELSE 0 END)                             AS OT_Employees
FROM hr_data
GROUP BY Department;
```
